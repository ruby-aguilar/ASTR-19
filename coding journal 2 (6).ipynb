```python
import numpy as np
import matplotlib.pyplot as plt
```

This pulls 1000 random numbers distributed logistically

```python
x = np.random.logistic(0, 1, 1000)
```

This creates a histogram of the values from above using 100 bins

```python
width = 0.5
histmin = np.floor(min(x))
histmax = np.ceil(max(x)) + width

plt.hist(x, bins = 100, alpha = 0.5, edgecolor = 'black')
plt.xlabel('n per bin')
plt.ylabel('x')

plt.savefig('session_4_of_coding_journal_2.pdf', bbox_inches = 'tight', facecolor = 'white')
```
